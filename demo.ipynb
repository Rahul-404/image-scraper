```python
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.content, "html.parser")
```

```python
image_tags = soup.find_all("img")
```

```python
del image_tags[0]
```

```python
# create folder to keep images
save_dir = "image/"

if not os.path.exists(save_dir):
    os.mkdir(save_dir)
else:
    print(1)
```

```python
requests.get(image_tags[0]["src"]).content
```

```python
for i in image_tags:
    print(i["src"])
    image_url = i["src"]
    image_data = requests.get(image_url).content

    with open(os.path.join(save_dir, f"{query}_{image_tags.index(i)}.jpeg"), "wb") as f:
        f.write(image_data)
```

```python
# delete all namo images
image_names = [i for i in os.listdir(save_dir) if i.startswith(query)]

for i in image_names:
    os.remove(os.path.join(save_dir, i))
```
